```python
import json
import pandas as pd
```

```python
json_file_path = "data\crime\crimes_category.json"

with open(json_file_path, 'r') as jsonfile:
    data_category = json.load(jsonfile)

print(data_category)

csv_file_path = "data\crime_data.csv"

data_complete = pd.read_csv(csv_file_path, names=["Departement", "Annee", "Mois", "Type_de_crime", "Nb_crimes"])

data_complete
```

```python
def find_key_by_value(dictionary, target_value):
    for key, values in dictionary.items():
        for value in values:
            if value == target_value:
                return key
    return target_value

new_col = []

for infraction in data_complete.values:
    new_col.append(find_key_by_value(data_category, infraction[3]))

new_col
```

```python
data_complete["Type_de_crime"] = new_col

data_complete['Annee'] = data_complete['Annee'].astype(str)

df_grouped = data_complete.groupby(['Departement', 'Annee', 'Mois', 'Type_de_crime']).agg({'Nb_crimes': 'sum'}).reset_index()
df_grouped
```

```python
df_grouped.to_csv('data/crime_data_clean.csv', index=False)
```
