```python
import pandas as pd
import json
import os

month_names = {
    2: 'Janvier',
    3: 'Février',
    4: 'Mars',
    5: 'Avril',
    6: 'Mai',
    7: 'Juin',
    8: 'Juillet',
    9: 'Août',
    10: 'Septembre',
    11: 'Octobre',
    12: 'Novembre',
    13: 'Décembre'
}

# Emplacement du répertoire contenant les fichiers Excel
dir_path = './data/crime/'

# Dictionnaire pour stocker les données
crime_data = {}

# Parcourir chaque fichier Excel dans le répertoire
for file_name in os.listdir(dir_path):
    if file_name.endswith('.xlsx'):
        # Extraire l'année du nom du fichier
        year = file_name.split('.')[0]
        
        # Lire le fichier Excel
        xlsx_file = pd.ExcelFile(os.path.join(dir_path, file_name))
        
        # Parcourir chaque feuille (mois) dans le fichier Excel
        for month in range(2,14):
            df = pd.read_excel(xlsx_file, sheet_name=month, header=2)
            
            # Traiter la feuille de calcul pour chaque département
            for department in df.columns[1:]:  # On commence à 1 pour exclure 'Index de l'État 4001'
                # Vérifier et initialiser le chemin dans le dictionnaire
                if department not in crime_data:
                    crime_data[department] = {}
                if year not in crime_data[department]:
                    crime_data[department][year] = {}
                if month_names[month] not in crime_data[department][year]:
                    crime_data[department][year][month_names[month]] = {}

                # Itérer sur chaque ligne pour capturer les données des crimes
                for index, row in df.iterrows():
                    crime_type = row['Index de l\'Etat 4001']
                    crime_count = row[department]
                    crime_data[department][year][month_names[month]][crime_type] = crime_count

# Convertir le dictionnaire en JSON
json_output = json.dumps(crime_data, indent=2, ensure_ascii=False)

# Sauvegarder le JSON dans un fichier
output_file = 'data/crime_data.json'
with open(output_file, 'w', encoding='utf-8') as f:
    f.write(json_output)

# Afficher la sortie pour vérification
print(json_output[:500])  # Afficher les 500 premiers caractères pour vérifier
```
